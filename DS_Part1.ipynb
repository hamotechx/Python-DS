# Python Data Structure for Beginners

```python
#1 List in python 
#Lists are the most flexible data structure in Python,and commonly used to store a collection of items.
#Lists are ordered,mutable,and allow duplicate elements.   
#Lists are defined using square brackets [] and elements are separated by commas.
#Use Cases: storing collections of items,manipulating data,sequences of elements.
#Creating a List
#Example:
my_list=[1,2,3,'apple','banana',True,3.14]
print("Original List:",my_list)
#Accessing List Elements by Index
list_access=my_list[3]
print("Accessed Element:",list_access)
#Modifying List Elements    
my_list[1]='orange'
print("Modified List:",my_list)
#Adding Elements to a List
my_list.append('grape') #append() method adds an element to the end of the list
print("List after Append:",my_list)
my_list.insert(2,'kiwi') #insert() method inserts an element at a specified position
print("List after Insert:",my_list)
#Removing Elements from a List
my_list.remove('apple')  #remove() method removes the first occurrence of a specified element                                                             
print("List after Remove:",my_list)
popped_element=my_list.pop() #pop() method removes and returns the last element of the list
print("Popped Element:",popped_element)
print("List after Pop:",my_list)
#List Slicing
sliced_list=my_list[1:4] #slicing to get elements from index 1 to 3
print("Sliced List:",sliced_list)
#List Length
list_length=len(my_list) #len() function returns the number of elements in the list
print("Length of List:",list_length)
# Iterating through a List
print("Iterating through List:")
for item in my_list:
    print(item) 
#List Comprehension
squared_numbers=[x**2 for x in range(1,6)] #creating a list of squared numbers using list comprehension
print("Squared Numbers List:",squared_numbers)
#Summary
#Lists are versatile data structures in Python that allow you to store and manipulate collections of items easily. 
#They support various operations such as accessing, modifying, adding, and removing elements, as well as slicing and iterating through the list. 
#List comprehension provides a concise way to create new lists based on existing ones.
#Run the code line by line to see how lists work in Python!
#This script demonstrates the basic operations and functionalities of lists in Python.    
```

```python
#2 Tuples in Python
#Tuples are similar to lists but are immutable,meaning their elements cannot be changed after creation.
#Tuples are defined using parentheses () and elements are separated by commas.
#Use Cases: storing fixed collections of items,returning multiple values from functions.
#Creating a Tuple
#Example:
my_tuple=(1,2,3,'apple','banana',True,3.14)
print("Original Tuple:",my_tuple)
#Accessing Tuple Elements by Index
tuple_access=my_tuple[3] #Accessing the element at index 3
print("Accessed Element:",tuple_access)
#Tuples are immutable,so we cannot modify elements directly.
#However,we can create a new tuple by concatenating existing tuples.    
new_tuple=my_tuple + ('grape',)
print("New Tuple after Concatenation:",new_tuple)
#Tuple Slicing
sliced_tuple=my_tuple[1:4] #slicing to get elements from index 1 to 3
print("Sliced Tuple:",sliced_tuple)
#Tuple Length
tuple_length=len(my_tuple) #len() function returns the number of elements in the tuple
print("Length of Tuple:",tuple_length)
# Iterating through a Tuple
print("Iterating through Tuple:") #Iterating means going through each element one by one
for item in my_tuple:
    print(item)
#Tuple Comprehension (using generator expression)
squared_numbers_tuple=tuple(x**2 for x in range(1,6)) 
print("Squared Numbers Tuple:",squared_numbers_tuple)
#Summary
#Tuples are immutable data structures in Python that allow you to store fixed collections of items. 
#They support operations such as accessing elements, slicing, and iterating through the tuple.
#Tuple comprehension can be achieved using generator expressions.
#Run the code line by line to see how tuples work in Python!
```

```python
#3 Sets in Python
#Sets are unordered collections of unique elements in Python.
#Sets are defined using curly braces {curly braces} or the set(parentheses) for empty set.
#Use Cases: storing unique items,performing mathematical set operations like union,intersection,difference.
#Creating a Set
#Example:
my_set={1,2,3,'apple','banana',True,3.14}
print("Original Set:",my_set)
#Adding Elements to a Set
my_set.add('grape') #add() method adds an element to the set
print("Set after Add:",my_set)
#Modifying Sets
my_set.update([4,5,'kiwi']) #update() method adds multiple elements to the set
print("Set after Update:",my_set)
#Removing Elements from a Set
my_set.remove('apple') #remove() method removes a specified element from the set
print("Set after Remove:",my_set)
discarded_element=my_set.discard('banana') #discard() method removes an element if it exists
print("Set after Discard:",my_set)  
#Set Length
set_length=len(my_set) #len() function returns the number of elements in the set
print("Length of Set:",set_length)
# Iterating through a Set
print("Iterating through Set:")
for item in my_set:
    print(item)
#Set Comprehension
squared_numbers_set={x**2 for x in range(1,6)} #creating a set of squared numbers using set comprehension
print("Squared Numbers Set:",squared_numbers_set)
#Summary
#Sets are unordered collections of unique elements in Python that allow you to store and manipulate data without duplicates. 
#They support various operations such as adding, modifying, and removing elements, as well as iterating through the set.
#Set comprehension provides a concise way to create new sets based on existing ones.
#Since set elements are unordered,you cannot access them by index, and you cannot perform slicing operations.
#Set elements has no order so no need to add new elemnts at specific positions. 
#Run the code line by line to see how sets work in Python!
#This script demonstrates the basic operations and functionalities of sets in Python.
```

```python
#4 Dictionaries in Python
#Dictionaries are unordered collections of key-value pairs in Python.
#Dictionaries are defined using curly braces {} with key-value pairs separated by commas.
#Use Cases: storing data with key-value relationships,fast lookups and retrievals.
#Creating a Dictionary
#Example:
my_dict={'name':'Alice','age':30,'city':'New York'}
print("Original Dictionary:",my_dict)
#Accessing Dictionary Values by Key
name_value=my_dict['name']
print("Accessed Value for 'name':",name_value)
#Modifying Dictionary Values
my_dict['age']=31
print("Modified Dictionary:",my_dict)
#Adding Key-Value Pairs to a Dictionary
my_dict['profession']='Engineer'
print("Dictionary after Adding Key-Value Pair:",my_dict)
#Removing Key-Value Pairs from a Dictionary
removed_value=my_dict.pop('city') #pop() method removes a key-value pair by key and returns the value
print("Removed Value for 'city':",removed_value)
print("Dictionary after Pop:",my_dict)
del my_dict['profession'] #del statement removes a key-value pair by key
print("Dictionary after Deletion:",my_dict)
#Dictionary Length
dict_length=len(my_dict) #len() function returns the number of key-value pairs in the dictionary
print("Length of Dictionary:",dict_length)
# Iterating through a Dictionary
print("Iterating through Dictionary:")
for key, value in my_dict.items():
    print(f"Key: {key}, Value: {value}")
#Dictionary Comprehension
squared_numbers_dict={x: x**2 for x in range(1,6)} #creating a dictionary of squared numbers using dictionary comprehension
print("Squared Numbers Dictionary:",squared_numbers_dict)
#Summary
#Dictionaries are unordered collections of key-value pairs in Python that allow you to store and manipulate data efficiently. 
#They support various operations such as accessing, modifying, adding, and removing key-value pairs, as well as iterating through the dictionary.
#Dictionary comprehension provides a concise way to create new dictionaries based on existing data. 
#Run the code line by line to see how dictionaries work in Python!
#This script demonstrates the basic operations and functionalities of dictionaries in Python.
```

Choosing the right Data Structure 

#  Choosing the Right Data Structure
#Choosing the right data structure is crucial for efficient data management and manipulation in Python.
#Here are some guidelines to help you choose the appropriate data structure based on your requirements:
#1. Use Lists when:
#- You need an ordered collection of items.
#- You require mutability (ability to change elements). 
#- You want to allow duplicate elements.
#Examples: storing sequences of items,manipulating data collections.
#2. Use Tuples when:
#- You need an ordered collection of items.
#- You require immutability (elements cannot be changed).
#- You want to store fixed collections of items.
#Examples: returning multiple values from functions,storing constant data.  
#3. Use Sets when:
#- You need an unordered collection of unique items.
#- You want to perform mathematical set operations like union,intersection,difference.
#Examples: storing unique items,removing duplicates from a collection.
#4. Use Dictionaries when:
#- You need to store data in key-value pairs.
#- You require fast lookups and retrievals based on keys.
#Examples: storing configuration settings,user profiles,lookup tables.
#Summary
#Choosing the right data structure depends on the specific requirements of your application.
#Consider factors such as order, mutability, uniqueness, and key-value relationships when making your choice.
#By selecting the appropriate data structure,you can optimize performance and simplify data manipulation in your Python programs

Nest and combination of data structures

```python
#1 List and Tuple Combination
#Lists and tuples can be combined to create complex data structures.
#Example: A list of tuples representing student names and their scores
students_scores=[('Alice',85),('Bob',90),('Charlie',78)]
print("Students and Scores:",students_scores)
#Accessing Data
for student, score in students_scores:
    print(f"Student: {student}, Score: {score}")
#Modifying Data
#Since tuples are immutable,we cannot change them directly.  
#Summary
#Combining lists and tuples allows for more complex data structures that can represent real-world data effectively.
#In this example,a list of tuples is used to store student names along with their scores. Each tuple contains a name and a score, and the list holds multiple such tuples.
#This structure allows for easy access and manipulation of student data, demonstrating how lists and tuples can
#work together to create more sophisticated data representations.
list_of_coordinates = [
    (10.1, 20.2),
    (35.5, 45.1),
    (-15.0, 80.0)
]
print(f"Coordinate: {list_of_coordinates}")
```

```python
#2 Dictionary with List Values
#Dictionaries can have lists as values to store multiple items under a single key.  
#Example: A dictionary representing a student's name and their list of grades
student_grades={'Alice':[85,90,78],'Bob':[88,92,80],'Charlie':[70,75,68]}
print("Student Grades:",student_grades)
#Accessing Data
for student, grades in student_grades.items():
    print(f"Student: {student}, Grades: {grades}")
#Modifying Data
student_grades['Alice'].append(95) #Adding a new grade for Alice
print("Updated Student Grades:",student_grades)
#Summary    
#Using lists as values in dictionaries allows for storing multiple related items under a single key.
#In this example,a dictionary is used to store student names as keys and their corresponding list of
#grades as values.This structure enables efficient organization and retrieval of student grade data.
#It also allows for easy modification, such as adding new grades to a student's list.
#This combination of dictionaries and lists provides a powerful way to manage and manipulate complex data sets.
#Run the code line by line to see how lists and dictionaries work together in Python!
#Example 2:
user_data = {
    "username": "Alice",
    "interests": ["coding", "reading", "hiking"],  # Value is a list
    "location_coords": (40.7128, -74.0060)         # Value is a tuple
}
```

```python
#3 List of Dictionaries
#A list can contain multiple dictionaries to represent a collection of related data items.
#Example: A list of dictionaries representing multiple students and their details
students=[
    {'name':'Alice','age':30,'city':'New York'},
    {'name':'Bob','age':25,'city':'Los Angeles'},
    {'name':'Charlie','age':28,'city':'Chicago'}
]
print("Students List of Dictionaries:",students)
#Accessing Data 
for student in students:
    print(f"Name: {student['name']}, Age: {student['age']}, City: {student['city']}")
#Modifying Data
students[0]['age']=31 #Updating Alice's age
print("Updated Students List:",students)
#Summary
#Using a list of dictionaries allows for storing multiple related data items in a structured manner.
#In this example,a list is used to hold dictionaries,each representing a student's details such as name,age,and city.
#This structure enables easy access and manipulation of individual student data while maintaining the overall collection.
#Run the code line by line to see how lists and dictionaries work together in Python!
#Example 2: 
products = [
    {"id": 1, "name": "Laptop", "price": 999.99},
    {"id": 2, "name": "Mouse", "price": 19.99},
    {"id": 3, "name": "Keyboard", "price": 45.50}
]
for product in products:
    print(f"Product ID: {product['id']}, Name: {product['name']}, Price: ${product['price']}")
```
